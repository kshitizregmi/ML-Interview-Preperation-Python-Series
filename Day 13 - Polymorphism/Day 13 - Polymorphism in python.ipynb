# Operator Overloading Using Dunders

```python
# we are trying to overload + operator
# In python, operator overloading is done using Dunders. 
```

# Dunders


```python
class Point:
    def __init__(self, x):
        self.x = x
    def __add__(self, p2):
        return self.x + p2.x
```

```python
p1 = Point(x=5)
p2 = Point(x=20)
p1 + p2
```
