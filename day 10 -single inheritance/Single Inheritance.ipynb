In Python, inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class (called a child class or subclass) to inherit properties and behaviors (methods and attributes) from another class (called a parent class or superclass).



## Key Concepts
* Parent class (superclass): The class being inherited from.
* Child class (subclass): The class that inherits from the parent class.

## Why use inheritance?
* Reuse code (avoid duplication)
* Create a logical class hierarchy
* Extend or modify behaviors of the parent class


# What is Single Inheritance?
Single Inheritance is a type of inheritance in Python where a child class inherits from only one parent class. It allows the child class to reuse the methods and properties of the parent class.

```python
class Mammal:
    has_hair = True # class attribute
```

```python
class Human:
    number_of_legs = 2
```

```python
h1 = Human()
h1.number_of_legs
```

```python
h1.has_hair
```

```python
class Mammal: # parent class
    has_hair = True # class attribute
    def display_class_attribute(self):  # method of class Mammal
        print(f"has_hair={self.has_hair}")

class Human(Mammal):  # child class
    number_of_legs = 2
    def __init__(sef):
        super().__init__()
```

```python
h1 = Human()
h1.has_hair
```

```python
h1.display_class_attribute()
```
